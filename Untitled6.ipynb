```python
# Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import davies_bouldin_score

# Load the datasets
customers = pd.read_csv('Customers.csv')
transactions = pd.read_csv('Transactions.csv')

# Display the first few rows of each dataset
print("Customers Data:")
print(customers.head())
print("\nTransactions Data:")
print(transactions.head())

# Data Preprocessing
# Merge datasets on CustomerID
data = pd.merge(customers, transactions, on='CustomerID')

# Aggregate transaction data for each customer
transaction_summary = data.groupby('CustomerID').agg({
    'Quantity': 'sum',
    'TotalValue': 'sum',
    'Price': 'mean',
    'TransactionDate': 'count'
}).reset_index()

# Rename columns for clarity
transaction_summary.columns = ['CustomerID', 'TotalQuantity', 'TotalValue', 'AveragePrice', 'TransactionCount']

# Merge the aggregated transaction summary back with the customer data
customer_data = pd.merge(customers, transaction_summary, on='CustomerID')

# Feature Selection
features = customer_data[['TotalQuantity', 'TotalValue', 'AveragePrice', 'TransactionCount']]

# Normalize the features
scaler = StandardScaler()
normalized_features = scaler.fit_transform(features)

# Clustering
# Choose the number of clusters
k = 4  # You can change this value based on your analysis
kmeans = KMeans(n_clusters=k, random_state=42)
customer_data['Cluster'] = kmeans.fit_predict(normalized_features)

# Calculate clustering metrics
db_index = davies_bouldin_score(normalized_features, customer_data['Cluster'])

# Visualization
plt.figure(figsize=(10, 6))
sns.scatterplot(x='TotalValue', y='TotalQuantity', hue='Cluster', data=customer_data, palette='deep', s=100)
plt.title('Customer Segmentation based on Total Value and Quantity')
plt.xlabel('Total Value')
plt.ylabel('Total Quantity')
plt.legend(title='Cluster')
plt.show()

# Print results
print(f'Number of clusters: {k}')
print(f'Davies-Bouldin Index: {db_index}')

# Save the final customer data with cluster assignments
customer_data.to_csv('Customer_Segmentation_Results.csv', index=False)
print("Customer segmentation results saved to 'Customer_Segmentation_Results.csv'.")
```
